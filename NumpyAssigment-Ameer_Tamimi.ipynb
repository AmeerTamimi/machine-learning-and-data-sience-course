```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shape = (1000, 1000, 3)
```

```python
# Red image
height = 10
width = 10
channels = 3
red_arr = np.zeros(shape, dtype=np.uint8)
red_arr[:, :, :] = [255, 0, 0]
plt.imshow(red_arr)
plt.title("A Red Image")
plt.show()
```

```python
# Green image
green_arr = np.zeros(shape, dtype=np.uint8)
green_arr[:, :, :] = [0, 255, 0]
plt.imshow(green_arr)
plt.title("A Green Image")
plt.show()
```

```python
# output 1
height = 10
width = 10
channels = 3
combined_arr = np.concatenate((red_arr, green_arr), axis=0)
plt.axis('off')
plt.imshow(combined_arr)
```

```python
# output 2
combined_arr = np.concatenate((red_arr, green_arr), axis=1)
plt.axis('off')
plt.imshow(combined_arr)
```

```python
# output 3
combined_arr1 = np.concatenate((red_arr, green_arr), axis=1)
combined_arr2 = np.concatenate((green_arr, red_arr), axis=1)
combined_arr3 = np.concatenate((combined_arr1, combined_arr2), axis=0)
plt.axis('off')
plt.imshow(combined_arr3)
```

```python
# output 4
height = 10
width = 10
channels = 3

striped_arr = np.zeros((height, width, channels), dtype=np.uint8)
striped_arr[::2, :, :] = [255, 0, 0]
striped_arr[1::2, :, :] = [0, 255, 0]


plt.imshow(striped_arr)
plt.axis('off')
plt.show()
```

```python
# output 5
height = 10
width = 10
channels = 3

striped_arr = np.zeros((height, width, channels), dtype=np.uint8)
striped_arr[:, ::2, :] = [255, 0, 0]
striped_arr[:, 1::2, :] = [0, 255, 0]


plt.imshow(striped_arr)
plt.axis('off')
plt.show()
```
