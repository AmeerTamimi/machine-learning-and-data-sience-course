```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
list_data = [[2,5, "square", 50],
             [2,3, "square", 56],
             [3,2, "square", 36],
             [3,4, "square", 80],
             [6,3, "square", 70],
             [1,5, "square", 85],
             [4,8, "square", 60],
             [5,5, "square", 83],
             [7,6, "triangle", 77],
             [6,7, "triangle", 20],
             [7,5, "triangle", 36],
             [5,6, "triangle", 56],
             [9,7, "triangle", 40],
             [8,8, "triangle", 37],
             [8,7, "triangle", 55],
             [6,9, "triangle", 41],
             ]

df = pd.DataFrame(data = list_data, columns=["x", "y", "label", "pointValue"])
```

***1- Write python code to visualize data as the following image***

```python
# find and store square & triange value from label variable in  two  new variable
square_df = df[df["label"] == "square"]
triangle_df = df[df["label"] == "triangle"]
# scutter plot for square label
plt.scatter(square_df["x"], square_df["y"],
            s=square_df["pointValue"],
            c="blue", marker="s",
            label="Square")
# scutter plot for Triangle label
plt.scatter(triangle_df["x"], triangle_df["y"],
            s=triangle_df["pointValue"],
            c="orange", marker="^",
            label="Triangle")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Scatter plot with square & triangle markers")
plt.show()
```

**2- based on understanding of the KNN ML algorithm**

*2.1- Set K value as input from user after run a program & 2.2- Add new test value as input ( x, y )*

```python
45# Taking input from the user
del K
K = int(input("Enter K value: "))
x_new = float(input("Enter new X: "))
y_new = float(input("Enter new Y: "))
```

*2.3- Calculate the distance between new input and all points in Data-Frame*

```python
df["distance"] = np.sqrt((df["x"] - x_new)**2 + (df["y"] - y_new)**2)
```

```python
# show first five rows after added distance variable
df.head()
```

*2.4- Sort distances*

```python
df_sorted = df.sort_values(by="distance")
df_sorted
```

```python
df_sorted.describe()
```

*2.5- Select K values that have a minimum distance with a new input value*

```python
neighbors = df_sorted.head(K)
neighbors
```

*2.6- Detect the label for a new input data*

```python
new_label = neighbors["label"].mode()[0]
new_label
```

*2.7- Detect the pointValue for a new input data*

```python
new_value = int(neighbors["pointValue"].mean())
```

*2.8- Insert the new inoput and the label and the valuePoint into Data-Frame*

```python
del new_point
new_point = pd.DataFrame([[x_new, y_new, new_label, new_value, 0]],
                         columns=["x", "y", "label", "pointValue", "distance"])
df2 = pd.concat([df, new_point], ignore_index=True)
```

2.9- Visualize the new point with green color and suitlable mark and plot lines between K nearest neighbors and with new input

```python
# Visualization
fig, axs = plt.subplots(1, 4, figsize=(16, 4))

x_min, x_max = 0, 10
y_min, y_max = 0, 10

# step 1: Original data
axs[0].scatter(df2[df2["label"]=="square"]["x"], df2[df2["label"]=="square"]["y"],
               c="blue", marker="s", label="Square")
axs[0].scatter(df2[df2["label"]=="triangle"]["x"], df2[df2["label"]=="triangle"]["y"],
               c="orange", marker="^", label="Triangle")

axs[0].set_xlim(x_min, x_max)
axs[0].set_ylim(y_min, y_max)
axs[0].set_title("1- Original data")

# step 2: Position of new data point
axs[1].scatter(df2[df2["label"]=="square"]["x"], df2[df2["label"]=="square"]["y"],
               c="blue", marker="s")
axs[1].scatter(df2[df2["label"]=="triangle"]["x"], df2[df2["label"]=="triangle"]["y"],
               c="orange", marker="^")
axs[1].scatter(x_new, y_new, c="green", marker="x", s=100)

axs[1].set_xlim(x_min, x_max)
axs[1].set_ylim(y_min, y_max)
axs[1].set_title("2- position of new data point")

# step 3: Draw lines to K nearest neighbors
axs[2].scatter(df2[df2["label"]=="square"]["x"], df2[df2["label"]=="square"]["y"],
               c="blue", marker="s")
axs[2].scatter(df2[df2["label"]=="triangle"]["x"], df2[df2["label"]=="triangle"]["y"],
               c="orange", marker="^")
axs[2].scatter(x_new, y_new, c="green", marker="x", s=100)

# رسم خطوط
for i, row in neighbors.iterrows():
    axs[2].plot([x_new, row["x"]], [y_new, row["y"]], "--", c="purple")
axs[2].set_xlim(x_min, x_max)
axs[2].set_ylim(y_min, y_max)
axs[2].set_title("3- draw lines with K nearest")

# step 4: Detect label of new data point
axs[3].scatter(df2[df2["label"]=="square"]["x"], df2[df2["label"]=="square"]["y"],
               c="blue", marker="s")
axs[3].scatter(df2[df2["label"]=="triangle"]["x"], df2[df2["label"]=="triangle"]["y"],
               c="orange", marker="^")
axs[3].scatter(x_new, y_new, c="green", marker="^" if new_label=="triangle" else "s",
               s=120)
axs[3].set_xlim(x_min, x_max)
axs[3].set_ylim(y_min, y_max)
axs[3].set_title("4- Detect label of new data point")

plt.tight_layout()
plt.show()
```
